# Seasonality

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

to see if there is a trend that will appear on a cyclical basis

**[1] Decomposition**

1) Trend: pattern

2) Seasonal: cyclical effects

3) Residual: error of prediction

```python
s_dec_additive = seasonal_decompose(df.market_value, model = "additive")
s_dec_additive.plot()
plt.show()
```

**[2] Naive Decomposition**

1) Additive: obsered = **trend + season + residual**

2) multiplicative: observed = **trend x seasonal x residual**

```python
s_dec_multiplicative = seasonal_decompose(df.market_value, model = "multiplicative")
s_dec_multiplicative.plot()
plt.show()
#suggests no seasonality
```

# **ACF**

```python
import statsmodels.graphics.tsaplots as sgt
```

```python
sgt.plot_acf(df.market_value, lags = 40, zero = False)
plt.title('ACF Market Value',size = 24)
plt.ylim(-0.5,1.1)
plt.show()
```

```python
sgt.plot_acf(df.wn, lags = 40, zero = False)
plt.title('ACF WN',size = 24)
plt.ylim(-0.05,0.05)
plt.show()
```

**=> no autocorrelation as all of the vertical lines are within the shaded area**

```python
sgt.plot_acf(df.rw, lags = 40, zero = False)
plt.title('ACF RW',size = 24)
plt.ylim(-1,1.1)
plt.show()
```

**=> no autocorrelation as all of the vertical lines are within the shaded area**
