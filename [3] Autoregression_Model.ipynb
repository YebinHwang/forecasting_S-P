### [1] AR Model 1

```python
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_predict
```

```python
model_ar = ARIMA(df.market_value, order = (1,0,0))
#1 indicates that the number of past values we wish to incorporate into the model
#0 differencing
#0 indicates that not taking any of the residual values into consideration
```

```python
result_ar = model_ar.fit()
```

```python
result_ar.summary()
```

* **p**: The number of lag observations included in the model, also called the lag order.
* **d**: The number of times that the raw observations are differenced, also called the degree of differencing.
* **q**: The size of the moving average window, also called the order of moving average.

```python
print(result_ar.forecast(1))
print(result_ar.forecast(2))
```

# **Log-Likelyhood LLR Test**

```python
from scipy.stats.distributions import chi2
```

```python
def LLR_test(mod_1, mod_2, DF=1):
    L1=mod_1.fit().llf
    L2=mod_2.fit().llf
    LR=(2*(L2-L1))
    P = chi2.sf(LR,DF).round(3)
    return P
```

```python
LLR_test(model_ar_2,model_ar_3)
```

```python
LLR_test(model_ar_3,model_ar_4)
```

```python
model_ar_4 = ARIMA(df.market_value, order = [4,0,0])
results_ar_4 = model_ar_4.fit()
print(results_ar_4.summary())
print("LLR test:" + str(LLR_test(model_ar_3, model_ar_4)))
```
